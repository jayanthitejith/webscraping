```python
titles = []
ratings = []
descriptions = []

for file_number in range(1,64):
    
    with open(f"html_scripts/{file_number}.html", 'r') as reader:
        html_string = reader.read()
        soup = BeautifulSoup(html_string, 'html.parser')

    

    for box in soup.find_all(class_ = "paper mb-1"):
        try:
            titles.append(box.find_all(class_ = "d-ib c-midnight-100 js-log-click")[0].text)    # titles
            # print(box.find_all(class_ = "d-ib c-midnight-100 js-log-click")[0].text)
        except IndexError:
            titles.append("No Title")
        
        try:
        
            ratings.append(box.find_all(class_ = "c-midnight-90")[0].text)    # ratings
        except IndexError:
            ratings.append("No Star Rating")
        try:
            descriptions.append(box.find_all(class_ = "product-listing__paragraph")[0].text)    # descriptions
        except IndexError:
            descriptions.append("No Descriptions Available")
```

```python
final_output = pd.DataFrame(data={"Company Name":titles, "Rating":ratings, "Product Description":descriptions})
print(final_output.shape)
final_output.head()
```
